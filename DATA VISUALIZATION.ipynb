# SOLID LINE STYLE

```python
x=[1,2,3,4]
y=[7,6,4,19]
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'g',linewidth=2,linestyle='solid')
plt.show()
```

 MARKER SIZE, TYPE AND COLOR

```python
x=[1,2,3,4]
y=[7,6,4,19]
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'g',marker ='d',markersize=5,markeredgecolor='red',linewidth=2,linestyle='solid')
plt.show()
```

# CREATING MULTIPLE BAR CHART

```python
a=[1,2,3,4]
b=[7,6,4,9]
c=[4,5,3,6]
x= np.arange(len(a))
plt.xlabel('x')
plt.ylabel('y')
plt.bar(x,a,color=['r'],width =0.35)
plt.bar(x+.35,b,color=['b'],width =0.35)
plt.bar(x+.70,c,color=['g'],width =0.35)
plt.show()
```

CREATING HORIZONTAL BAR CHART

```python
a=[1,2,3,4]
b=[7,6,4,9]
c=[4,5,3,6]
x= np.arange(len(a))
plt.xlabel('x')
plt.ylabel('y')
plt.barh(x,a,color=['r'])
plt.barh(x+.35,b,color=['b'],alpha=0.2)
plt.barh(x+.70,c,color=['g'],alpha=0.5)
plt.legend(["op = 0.2", "op = 0.5", "op = 1"])
plt.show()
```

# PIE CHART

```python
a=[100,120,80,20,40]
b=['c','d','r','b','f']
plt.pie(a,labels=b)
plt.show()
```

# DISPLAY PIE CHART IN PERCENTAGE VALUE

```python
a=[100,120,80,20,40]
b=['c','d','r','b','f']
plt.pie(a,labels=b,autopct="%3d%%")
plt.show()
```

# DISPLAY PIE CHART USING THE FLOAT

```python
a=[100,120,80,20,40]
b=['c','d','r','b','f']
plt.pie(a,labels=b,autopct="%6.2f")
plt.show()
```

# DISPLAY PIE CHART USING THE FLOAT PERCENTATGE


```python

a=[100,120,80,20,40]
b=['c','d','r','b','f']
plt.pie(a,labels=b,autopct="%6.2f%%")
plt.show()
```

# DISPLAY PIECHART USING DIFFERENT COLORS GIVEN BY THE USER

```python
a=[100,120,80,20,40]
b=['cat','dog','rabbit','bird','fowl']
c=['r','b','g','c','d']
plt.pie(a,colors=c,labels=b)
plt.show()
```

# DISPLAY PIECHART USING EXPLODE OPTION

```python
a=[100,120,80,20,40]
b=['c','d','r','b','f']
e=[0.2,0,0.3,0.4,0]
plt.pie(a,labels=b,shadow =True,explode =e,autopct="%6.2f%%")
plt.show()
```
